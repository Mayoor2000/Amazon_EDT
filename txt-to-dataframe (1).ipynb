Function to convert training data of txt format into python DataFrame using **Pandas** and **numpy** library

```python
def Trans(fp) :
    data =[]
    with open(fp,"r") as reader :
        for line in reader.readlines():
            data.append(line)
    data2=[]
    import regex as re
    for i in data:
        data2.append(re.sub('\s+',' ', i))  

    data3=pd.DataFrame(data2)
    data3 = data3[0].str.split(" ",expand=True)
    data3[0]=data3[0].replace(['Weather','Time_taken'],['Weather_conditions','Time_taken_(min)'])
    data5=data3.copy()
    data5[1][data5[0]== 'Weather_conditions'] = data5[2]
    data5[1][data5[0] == 'Time_taken_(min)'] = data5[2]
    data5.drop([2,3],axis=1,inplace=True)
    data5.drop(20,inplace=True)
    dfc=data5.transpose()
    dfc.columns=dfc.iloc[0]
    dfc.drop(0,inplace=True)

    return dfc
```

First we will add file[0] to the DataFrame


```python
file="../input/txt-data/dataset/train/0.txt"
x=Trans(file)

```

Now iteratively we will append all the files for training dataset

```python
for i in range(1,45594):
    filepath="../input/txt-data/dataset/train/{0}.txt".format(i)
    y=Trans(filepath)
    x=pd.concat([x,y],ignore_index=True)
```

```python
x.to_csv('mycsvfile.csv',index=False)
```
